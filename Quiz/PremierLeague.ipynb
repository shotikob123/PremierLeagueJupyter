```python
#ვახდენ საჭირო ბიბლიოთეკების დაიმპორტებას
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#ვკითხულობ csv ფაილს
data = pd.read_csv('stats.csv')
```

```python
#ეკრანზე გამომაქ ეს წაკითხული csv ფაილი
data
```

```python
#ვირჩევ სასურველ სვეტებს სახელით და ვიღებ ამ სვეტების 100 სტრიქონს
selected_columns = ['team', 'wins', 'losses', 'goals', 'ontarget_scoring_att', 'total_cross', 'corner_taken', 'touches', 'clearance_off_line', 'season']
selected_data = data.iloc[:100][selected_columns]
```

```python
#გამომაქ ეკრანზე
selected_data
```

```python
#ვაკეთებ team სვეტის ინდექსირებას
selected_data.set_index('team', inplace=True)
```

```python
#გამომაქ ეკრანზე
selected_data
```

```python
#ვფილტრავ მონაცემებს მოგებების და წაგებების მიხედვით
filter_condition = (selected_data['wins'] >12.5) & (selected_data['losses'] < 7)
filtered_data = selected_data[filter_condition]
```

```python
#გამომაქ ეკრანზე
filtered_data
```

```python
#ვსორტავ მოგებების და გოლების მიხედვით
sorted_data = filtered_data.sort_values(by=['wins', 'goals'], ascending=False)
sorted_data
```

```python
#ვითვლი საშუალოს, სტანდარტულ გადახრას, მედიანას, მინიმალურს და მაქსიმალურს:
mean_wins = sorted_data['wins'].mean()
std_wins = sorted_data['wins'].std()
median_ontarget_scoring_att = sorted_data['ontarget_scoring_att'].median()
min_losses = sorted_data['losses'].min()
max_losses = sorted_data['losses'].max()
```

```python
print(mean_wins)
print(std_wins)
print(median_ontarget_scoring_att)
print(min_losses)
print(max_losses)
```

```python
# Bar chart გუნდების და მათი გამარჯვებების მიხედვით
plt.bar(sorted_data.index, sorted_data['wins'])
plt.xlabel('team')
plt.ylabel('wins')
plt.title('Teams & Wins')

# Line chart გოლების და სეზონების მიხედვით
fig, ax2 = plt.subplots(figsize=(10, 5))
sorted_data.groupby('season')['goals'].mean().plot(kind='line', ax=ax2, marker='o', color='green')
ax2.set_title('Average Number of Goals per Season')
ax2.set_ylabel('Goals')
ax2.set_xlabel('Season')

plt.show()

```
